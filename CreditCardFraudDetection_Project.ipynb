### The data is highly imbalanced
This data is imbalanced because more than 99% of data is present in one class (0) i.e. normal transactions. Thus, we cannot feed this data and train the machine learning model because it will not be able to recognise the fraudelent data as the data point is very low in this class (1).

```python
## Separating the data based on Legit (0) and Fraudelent(1) transaction for data analysis ##
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
legit.head()
```

Under-sampling:
Since it is highly imbalanced data and is unfit to go ahead. So here using Under-Sampling technique to correct the imbalanced data.


Now building a sample data from the original data for Normal/Legit and Fraudelent txns.

No of Fraudelent txns - 492

No of Normal/Legit txns - 284315

So, now from these 284315 normal txns 492 random txns will be taken out and then the ratio of fraudelent and normal txns will be 50:50 which will be better for training the machine learning model.

```python
### Getting 492 random records from the Legit records ###
legit_sample = legit.sample(n=492)
legit_sample.shape
```

```python
## Concatenating the fraud and legit_sample tables ###

new_df = pd.concat([legit_sample, fraud], axis = 0)
new_df.shape
```

Here, axis = 0 means rows and axis = 1 means columns. So, here the the records will get added row-wise 1 by 1 and not column-wise.

```python
new_df.head()
```

```python
new_df.tail()
```

So here as legit_sample was put first into concat hence that is the one on top of this table (new_df).

```python
new_df['Class'].value_counts()
```

So here we can see now our data is uniformly distributed i.e. 492 txns each for legit and fraudulent class.

```python
new_df.groupby('Class').mean()
```

Here after the comparing the mean of the legit txns the values are very much similar. This means the property/characteristic of the legit txns is same which will help the machine learning model to differentiate between the legit and fraud txn.

This step is important which helps us check whther the samples pulled are right for going ahead or not.

```python
### Splitting the data into features(X) and target(Y) ###
X = new_df.drop(columns = 'Class', axis = 1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
## Splitting the data into training & testing data ##
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify = Y , random_state= 2)
```

Here, the train test split is 8:2. 

sratify(Y) based on Y will evenly distribute the data, otherwise the Y - 0 & 1 distribution will be very different in the training and testing data.

random_state = 2 ==> This helps control the random shuffling of the data. This means that if the same 'random_state' value is used in multiple runs of the train-test split function, you will get the same split of data each time.

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model on X_train & Y_train. Logistic Regression is used here because for binary classification its the best suitable model. Other models can be used too but accuracy will be different.

```python
### Training the model ###

model = LogisticRegression() ## creating Logistic regression object

model.fit(X_train, Y_train)  ## Training the model on training data
```

```python
### Model Evaluation based on accuracy score ###

## Accuracy on training data

X_train_prediction = model.predict(X_train) ## the predicted values for X_train will get stored here

Accuracy_training_data = accuracy_score(X_train_prediction, Y_train) ##predicted value will get compared with Y_train(original values)
```

```python
print('Training data accuracy:', Accuracy_training_data)
```

```python
## Accuracy on test data ##

X_test_prediction = model.predict(X_test) ## the predicted values for X_test will get stored here

Accuracy_test_data = accuracy_score(X_test_prediction, Y_test) ## comparison b/w Y_test and X_test_prediction (predicted values)

print('Test data accuracy:', Accuracy_test_data)
```

Here, this model is perfect and the accuracy score of training and testing data is almost same.

Overfitting - If Train_accuracy_score > Test_accuracy_score --- Where model gets overtrained on training data

Underfitting - If Train_accuracy_score < Test_accuracy_score --- Where the model gets underfit data.
