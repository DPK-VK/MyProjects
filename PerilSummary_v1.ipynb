```python
concatenated_df_LOC = pd.DataFrame()
concatenated_df_USD = pd.DataFrame()
concatenated_df_EUR = pd.DataFrame()
```

```python
# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):  # Assuming all files are Excel files
        file_path = os.path.join(folder_path, filename)
        
        # Check if the sheet 'Aggregate Data' exists in the Excel file
        excel_sheets = pd.ExcelFile(file_path).sheet_names
        if 'Aggregate Data' in excel_sheets:
            # Read the Excel file, considering row 6 as column headers and starting from row 7
            excel_data = pd.read_excel(file_path, sheet_name='Aggregate Data', header=5)
            
            # Concatenate the data to the main DataFrame
            concatenated_df_LOC = pd.concat([concatenated_df_LOC, excel_data])
```

```python
concatenated_df_LOC.reset_index(drop=True, inplace=True)
```

```python
concatenated_df_LOC.dtypes
```

```python
# Replacing '-' to 0 in BI Value
concatenated_df_LOC['BI Value'] = concatenated_df_LOC['BI Value'].replace('-',0)
```

```python
# Converting BI Value column to float
concatenated_df_LOC['BI Value'] = concatenated_df_LOC['BI Value'].astype(float)
```

```python
# Iterate through each file in the folder for Aggregate Data USD
for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):  # Assuming all files are Excel files
        file_path = os.path.join(folder_path, filename)
        
        # Check if the sheet 'Aggregate Data' exists in the Excel file
        excel_sheets = pd.ExcelFile(file_path).sheet_names
        if 'Aggregate Data USD' in excel_sheets:
            # Read the Excel file, considering row 6 as column headers and starting from row 7
            excel_data = pd.read_excel(file_path, sheet_name='Aggregate Data USD', header=5)
            
            # Concatenate the data to the main DataFrame
            concatenated_df_USD = pd.concat([concatenated_df_USD, excel_data])
```

```python
# Replacing '-' to 0 in BI Value
concatenated_df_USD['BI Value'] = concatenated_df_USD['BI Value'].replace('-',0)

# Converting BI Value column to float
concatenated_df_USD['BI Value'] = concatenated_df_USD['BI Value'].astype(float)
```

```python
concatenated_df_USD.reset_index(drop=True, inplace=True)
concatenated_df_USD.head()
```

```python
concatenated_df_USD.dtypes
```

```python
# Iterate through each file in the folder for Aggregate Data EUR
for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):  # Assuming all files are Excel files
        file_path = os.path.join(folder_path, filename)
        
        # Check if the sheet 'Aggregate Data' exists in the Excel file
        excel_sheets = pd.ExcelFile(file_path).sheet_names
        if 'Aggregate Data EUR' in excel_sheets:
            # Read the Excel file, considering row 6 as column headers and starting from row 7
            excel_data = pd.read_excel(file_path, sheet_name='Aggregate Data EUR', header=5)
            
            # Concatenate the data to the main DataFrame
            concatenated_df_EUR = pd.concat([concatenated_df_EUR, excel_data])
```

```python
# Replacing '-' to 0 in BI Value
concatenated_df_EUR['BI Value'] = concatenated_df_EUR['BI Value'].replace('-',0)

# Converting BI Value column to float
concatenated_df_EUR['BI Value'] = concatenated_df_EUR['BI Value'].astype(float)
```

```python
concatenated_df_EUR.reset_index(drop=True, inplace=True)
concatenated_df_EUR.head()
```

```python
concatenated_df_LOC['SUM_3'] = concatenated_df_LOC['Buildings / Motor Value'] + concatenated_df_LOC['Contents Value'] + concatenated_df_LOC['BI Value']
concatenated_df_USD['SUM_3'] = concatenated_df_USD['Buildings / Motor Value'] + concatenated_df_USD['Contents Value'] + concatenated_df_USD['BI Value'] 
concatenated_df_EUR['SUM_3'] = concatenated_df_EUR['Buildings / Motor Value'] + concatenated_df_EUR['Contents Value'] + concatenated_df_EUR['BI Value']
```

```python
# Filter rows based on 'Property LOB' values 'PRO_RES' and 'PRO_COM'
pro_res_df = concatenated_df_LOC[concatenated_df_LOC['Property LOB'] == 'PRO_RES']
pro_com_df = concatenated_df_LOC[concatenated_df_LOC['Property LOB'] == 'PRO_COM']
pro_ind_df = concatenated_df_LOC[concatenated_df_LOC['Property LOB'] == 'PRO_IND']
```

```python
# Group by 'Country', 'Peril', 'Currency', and 'Property LOB', and calculate sums of 'BI Value'
pro_res_summary = pro_res_df.groupby(['Country', 'Peril', 'Currency']).agg({'SUM_3': 'sum'}).reset_index()
pro_com_summary = pro_com_df.groupby(['Country', 'Peril', 'Currency']).agg({'SUM_3': 'sum'}).reset_index()
pro_ind_summary = pro_ind_df.groupby(['Country', 'Peril', 'Currency']).agg({'SUM_3': 'sum'}).reset_index()
```

```python
pro_com_summary.head()
```

```python
# Merge the two summaries on 'Country', 'Peril', and 'Currency'
summary_df1 = pd.merge(pro_res_summary, pro_com_summary, on=['Country', 'Peril', 'Currency'], suffixes=('_Residential', '_Commercial'))
summary_df = pd.merge(summary_df1, pro_ind_summary, on=['Country', 'Peril', 'Currency'], how = 'outer')

# Calculate 'Property Residential Amount' and 'Property Commercial Amount'
summary_df['Property Residential Amount'] = summary_df['SUM_3_Residential']
summary_df['Property Commercial Amount'] = summary_df['SUM_3_Commercial']
summary_df['Property Industries Amount'] = summary_df['SUM_3']

#Replacing NA for industries
summary_df['Property Industries Amount'].fillna(0, inplace=True)

# Calculate 'Total'
summary_df['Total'] = summary_df['Property Residential Amount'] + summary_df['Property Commercial Amount'] + summary_df['Property Industries Amount']

# Reorder columns as per requirement
summary_df = summary_df[['Country', 'Peril', 'Currency', 'Property Residential Amount', 'Property Commercial Amount', 'Property Industries Amount','Total']]

# Display the summary DataFrame
print(summary_df)
```
